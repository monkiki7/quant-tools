```python
import pandas as pd
import numpy as np
import tushare as ts
import matplotlib.pyplot as plt
```

```python
stock_data = ts.get_k_data('600848',start='2014-03-01')
```

```python
ma_list = [5, 20, 60]
```

```python
# 简单算术平均线
for ma in ma_list:
    stock_data['MA_'+str(ma)] = stock_data['close'].rolling(center=False, window=ma).mean()
# 计算指数平滑移动平均线EMA
for ma in ma_list:
    stock_data['EMA_'+str(ma)] = stock_data['close'].ewm(ignore_na=False,adjust=True,min_periods=0,span=ma).mean()
```

```python
# 处理index
time_index = pd.to_datetime(stock_data['date'])
stock_data.set_index(time_index, inplace=True)
```

```python
stock_data['close'].plot(figsize=(20,15))
for ma in ma_list:
    stock_data['MA_'+str(ma)].plot()
    #stock_data['EMA_'+str(ma)].plot()
plt.legend()
plt.grid(True)
plt.show()
```
